### Sherlock Agent Workspace
Environment is perfectly synced with `backend/.env` and `backend/requirements.txt`.

```python
import os
from dotenv import load_dotenv

load_dotenv()


print("Database Host:", os.getenv("POSTGRES_HOST"))
print("Model ID:     ", os.getenv("BEDROCK_MODEL_ID"))
```

```python
from langchain_aws import ChatBedrockConverse
from langchain_core.messages import HumanMessage


model = ChatBedrockConverse(
    model_id="moonshotai.kimi-k2.5",
    region_name="ap-south-1"
)

print("⏳ Asking Kimi for a poem\n")
response = model.invoke([HumanMessage(content="Write a short poem about engineering projects.")])
print(response.content)
```
