# Model Invocation Test
This notebook demonstrates how to load your `.env` configuration and invoke an Amazon Bedrock model using LangChain.

```python
import os
from dotenv import load_dotenv
from langchain_aws import ChatBedrock

# 1. Load environment variables (this reads the symlinked notebooks/.env)
load_dotenv()

print("Environment loaded!")
print(f"Target Model: {os.getenv('BEDROCK_MODEL_ID')}")
```

```python
# 2. Initialize the Bedrock LangChain Wrapper
llm = ChatBedrock(
    model_id=os.getenv("BEDROCK_MODEL_ID"),
    region_name=os.getenv("AWS_DEFAULT_REGION", "us-east-1")
)

# Note: The AWS_ACCESS_KEY_ID and AWS_SECRET_ACCESS_KEY from your .env
# are automatically picked up by the underlying boto3 client.
```

```python
# 3. Test an invocation
response = llm.invoke("Hello! Are you operational?")

print(response.content)
```
